## 2 System Size Trends

### 2.1 Queries

Here, we create an OEDIAthena object and use it to run sql queries on our data lake in AWS. Specifically, we calculate the median as well as a percentile range on system size by year. Note that in tracking-the-sun-2021, the "NON-RES" category is for records where a more specific customer-segment could not be determined.

```python
from oedi.AWS.athena import OEDIAthena
athena = OEDIAthena(staging_location = S3_STAGING_DIR, region_name = AWS_REGION_NAME)

# Query for residential system sizes
query_string_res = f"""
    SELECT
        CAST(YEAR(installation_date) AS smallint) AS year,
        approx_percentile(system_size_dc,0.2) AS P20,
        approx_percentile(system_size_dc,0.5) AS P50,
        approx_percentile(system_size_dc,0.8) AS P80
    FROM {DATABASE_NAME}.{TABLE_NAME}
    WHERE system_size_dc > 0 AND
        CAST(YEAR(installation_date) AS smallint) BETWEEN 2000 AND 2020 AND
        customer_segment = 'RES'
    GROUP BY CAST(YEAR(installation_date) AS smallint)
    HAVING COUNT(*) > 20
    ORDER BY CAST(YEAR(installation_date) AS smallint);
    """
pv_size_res = athena.run_query(query_string_res)

# Query for non-residential system sizes
query_string_non = f"""
    SELECT
        CAST(YEAR(installation_date) AS smallint) AS year,
        approx_percentile(system_size_dc,0.2) AS P20,
        approx_percentile(system_size_dc,0.5) AS P50,
        approx_percentile(system_size_dc,0.8) AS P80
    FROM {DATABASE_NAME}.{TABLE_NAME}
    WHERE system_size_dc > 0 AND
        CAST(YEAR(installation_date) AS smallint) BETWEEN 2000 AND 2020 AND
        customer_segment IN ('COM', 'SCHOOL', 'GOV', 'NON-PROFIT', 'NON-RES')
    GROUP BY CAST(YEAR(installation_date) AS smallint)
    HAVING COUNT(*) > 20
    ORDER BY CAST(YEAR(installation_date) AS smallint);
    """
pv_size_non = athena.run_query(query_string_non)

pv_size_res.set_index("year", inplace = True)
pv_size_non.set_index("year", inplace = True)
```

## 2.2 Plot the data

We can make plots of this data using pyplot that are similar to those in the [emp report](https://emp.lbl.gov/sites/default/files/2_tracking_the_sun_2021_report.pdf), page 11.

```python
import numpy as np
import matplotlib.pyplot as plt

def tsplot(t, p20, p50, p80, color = 'green', ax = None):
    plt.sca(ax)
    plt.fill_between(t, p20, p50, alpha = 0.2, color = color, label = 'Percentile Range (20th to 80th)')
    plt.fill_between(t, p50, p80, alpha = 0.2, color = color)
    plt.plot(t, p50, color = color, label = 'Median')
    plt.ylabel('kW_DC')

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (8,8), sharex = True)
ax1.set_title("Residential PV Systems")
ax1.set_ylim(bottom = 0, top = 12)
ax2.set_title("Non-Residential PV Systems")
ax2.set_ylim(bottom = 0, top = 250)
ax2.set_xticks([2000, 2005, 2010, 2015, 2020])
fig.suptitle("System Size Trends", size = 16)

tsplot(np.array(pv_size_res.index, dtype = 'float'),
       np.array(pv_size_res.P20, dtype = 'float'),
       np.array(pv_size_res.P50, dtype = 'float'),
       np.array(pv_size_res.P80, dtype = 'float'),
       ax = ax1)

tsplot(np.array(pv_size_non.index, dtype = 'float'),
       np.array(pv_size_non.P20, dtype = 'float'),
       np.array(pv_size_non.P50, dtype = 'float'),
       np.array(pv_size_non.P80, dtype = 'float'),
       color = "orange",
       ax = ax2)

ax1.legend(loc = 'upper left');
```

## 3 PV System Cost by State

### 3.1 Queries

Next, let's analyze the cost of installing a residential solar system in 2020. First, we will use the OEDIAthena object created earlier to query the appropirate data.

```python
# Query for residential system prices by state
query_string = f"""
    SELECT
        state,
        approx_percentile(total_installed_price / system_size_dc, 0.2) / 1000 AS P20,
        approx_percentile(total_installed_price / system_size_dc, 0.5) / 1000 AS P50,
        approx_percentile(total_installed_price / system_size_dc, 0.8) / 1000 AS P80
    FROM {DATABASE_NAME}.{TABLE_NAME}
    WHERE
        system_size_dc > 0 AND
        total_installed_price > 0 AND
        CAST(YEAR(installation_date) AS smallint) = 2020 AND
        customer_segment = 'RES'
    GROUP BY state
    HAVING COUNT(*) > 20
    ORDER BY P50;
    """
pv_price = athena.run_query(query_string)
```

### 3.2 Plot the Data

Using pyplot again, we can generate a bar graph similar to the one in the [emp report](https://emp.lbl.gov/sites/default/files/2_tracking_the_sun_2021_report.pdf), page 28. Note that the emp report used additional data that were provided confidentially and were scrubbed out before publishing the public data. For example, that the bar graph generated below does not have a bar for WA state.

```python
fig, ax = plt.subplots(figsize = (10,4))
ax.set_ylabel('2020$ / W_dc')
fig.suptitle('Median Installed Price and 20th/80th Percentiles for Residential PV Systems', size = 16)
bar_plot = ax.bar(
    pv_price.state,
    pv_price.P50,
    yerr = (pv_price.P50 - pv_price.P20, pv_price.P80 - pv_price.P50),
    color = "green"
)
```
